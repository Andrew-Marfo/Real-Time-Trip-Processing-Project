## Importing packages

```python
import pandas as p
```

### Loading and Inspecting data

```python
# load the data
trip_start_path = "./../data/trip_start.csv"
trip_end_path = "./../data/trip_end.csv"

trip_start = p.read_csv(trip_start_path)
trip_end = p.read_csv(trip_end_path)
```

```python
# Inspecting trip start data
print("Trip Start Data:")
trip_start.head(5)
```

```python
# Inspecting trip end data
print("Trip End Data:")
trip_end.head(5)
```

```python
# Describing start trip the data
print("Start Trip Data Description:")
trip_start.describe(include='all')
```

```python
# Describing end trip data
print("End Trip Data Description:")
trip_end.describe(include='all')
```

### Transformations and KPI computations

```python
# merging the two dataframes
trips = trip_start.merge(trip_end, on='trip_id', how='inner')

# Inspecting the merged data
print("Merged Trip Data:")
trips.head(5)
```

```python
# Extracting dates from the datetime columns
trips["pickup_date"] = p.to_datetime(trips["pickup_datetime"]).dt.date
trips["dropoff_date"] = p.to_datetime(trips["dropoff_datetime"]).dt.date
```

#### KPI's Computations

```python
# total_fare: The sum of all fares collected from completed trips on a given day
trips_total_fare = trips.groupby("pickup_date")["fare_amount"].sum().reset_index()
trips_total_fare.rename(columns={"pickup_date": "date", "fare_amount":"total_fare"}, inplace=True)
trips_total_fare.head()
```

```python
# count_trips: The total number of completed trips for the day.
trips_count = trips.groupby("pickup_date")["trip_id"].count().reset_index()
trips_count.rename(columns={"pickup_date": "date", "trip_id": "total_trips"}, inplace=True)
trips_count.head()
```

```python
# average_fare: The average fare amount per trip for the day.
trips_avg_fare = trips.groupby("pickup_date")["fare_amount"].mean().reset_index()
trips_avg_fare.rename(columns={"pickup_date": "date", "fare_amount": "average_fare"}, inplace=True)
trips_avg_fare.head()
```

```python
# max_fare: The highest single fare recorded among completed trips for the day.
trips_max_fare = trips.groupby("pickup_date")["fare_amount"].max().reset_index()
trips_max_fare.rename(columns={"pickup_date": "date", "fare_amount": "max_fare"}, inplace=True)
trips_max_fare.head()
```

```python
# min_fare: The lowest fare among completed trips for the day
trips_min_fare = trips.groupby("pickup_date")["fare_amount"].min().reset_index()
trips_min_fare.rename(columns={"pickup_date": "date", "fare_amount": "min_fare"}, inplace=True)
trips_min_fare.head()
```

```python
# Joining all KPI's into one table sequentially
trips_daily_level_metrics = trips_total_fare.merge(trips_count, on="date", how="inner")
trips_daily_level_metrics = trips_daily_level_metrics.merge(trips_avg_fare, on="date", how="inner")
trips_daily_level_metrics = trips_daily_level_metrics.merge(trips_max_fare, on="date", how="inner")
trips_daily_level_metrics = trips_daily_level_metrics.merge(trips_min_fare, on="date", how="inner")

# Display the result
trips_daily_level_metrics.head()
```
